```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', 999)
```

```python
#read the data sets
csv_path = "advanced_stats.csv"
advanced_stats = pd.read_csv(csv_path)

csv_path = "allgames_stats.csv"
allgames_stats = pd.read_csv(csv_path)

csv_path = "allstar_games_stats.csv"
allstar_games_stats = pd.read_csv(csv_path)

csv_path = "game_highs_stats.csv"
game_highs_stats = pd.read_csv(csv_path)

csv_path = "per_game_stats.csv"
per_game_stats = pd.read_csv(csv_path)

csv_path = "salaries.csv"
salaries = pd.read_csv(csv_path)

csv_path = "totals_stats.csv"
totals_stats = pd.read_csv(csv_path)
```

```python
# rename all dataframe columns to make them more readable

dataframes = [allgames_stats, advanced_stats, allstar_games_stats, game_highs_stats, per_game_stats, totals_stats]

for i in dataframes:
    i.rename(columns = {'Tm':'Team', 
                            'Lg':'League', 
                            'Pos':'Position',
                            'G':'Game', 
                            'MP':'Minutes Played', 
                            'PER':'Player Efficiency Rating', 
                            'TS%':'True Shooting %', 
                            '3PAr':'Percent of Field Goal Attempts that were 3-Pointers', 
                            'FTr':'Free Throw Rate', 
                            'ORB%':'Offensive Rebound %', 
                            'DRB%':'Defensive Rebound %', 
                            'TRB%':'Total Rebounds', 
                            'AST%':'Assist %', 
                            'STL%':'Steal %', 
                            'BLK%':'Block %', 
                            'TOV%':'Turnover %', 
                            'USG%':'Usage %', 
                            'OWS':'Offensive Win Shares', 
                            'DWS':'Defensive Win Shares', 
                            'WS':'Win Shares', 
                            'WS/48':'Win Shares per 48 Min', 
                            'OBPM':'Offensive Box Plus/Minus', 
                            'DBPM':'Defensive Box Plus/Minus', 
                            'BPM':'Box Plus/Minus', 
                            'VORP':'Value Over Replacement Player', 
                            'RSorPO':'Rg. Season or Playoff'}, 
                             inplace=True)

moredataframes = [allgames_stats, allstar_games_stats, game_highs_stats, per_game_stats, totals_stats]

for i in moredataframes:
    i.rename(columns = { 'GS':'Games Started',
                               'FG':'Field Goals', 
                               'FGA':'Field Goals Attempted', 
                               'FG%':'Field Goal %', 
                               '3P':'3 Pointers', 
                               '3PA':'3 Pointers Attempted', 
                               '3P%':'3 Pointers %', 
                               'FT':'Free Throw',
                               'FTA':'Free Throws Attempted', 
                               'FT%':'Free Throw Percent', 
                               'ORB':'Offensive Rebound', 
                               'DRB':'Defensive Rebound',
                               'TRB':'Total Rebounds', 
                               'AST':'Assist', 
                               'STL':'Steals', 
                               'BLK':'Blocks', 
                               'TOV':'Turnover', 
                               'PF':'Personal Foul',
                               'PTS':'Points'},
                                inplace=True)

evenmoredataframes = [game_highs_stats, per_game_stats, totals_stats]

for i in evenmoredataframes:
    i.rename(columns = {'2P':'2 Pointers',
                        '2PA':'2 Pointers Attempted'},
                        inplace=True)

looktherearemoredataframes = [per_game_stats, totals_stats]

for i in looktherearemoredataframes:
    i.rename(columns = {'eFG%':'Effective Field Goal %',
                       '2P%':'2 Pointer %'}, 
                        inplace=True)
    
allgames_stats.rename(columns = {'Opp':'Opposing Team', 
                                 'X':'Home or Away'},
                                 inplace=True)
    
game_highs_stats.rename(columns = {'GmSc':'Game Score'}, inplace=True)
salaries.rename(columns={'Lg':'League'}, inplace=True)
```

```python
points_df = totals_stats[['Season', 'Points', 'Minutes Played', 'Player']]
points_df.head()
```

```python
pointcost_df = pd.merge(points_df, salaries, left_on=['Season','Player'], right_on = ['Season','Player'])[['Season', 'Points', 'Minutes Played', 'Salary', "Player"]]
pointcost_df.head()
```

```python
newsalary = []
for index,row in pointcost_df.iterrows():
    salary_int = int(row[3][1:])
    newsalary.append(salary_int)
```

```python
pointcost_df['Salary Int'] = newsalary
pointcost_df = pointcost_df.drop('Salary', axis=1)
pointcost_df = pointcost_df.rename(columns={'Salary Int':'Salary'})
pointcost_df
```

```python
bryant_pointcost_df = pointcost_df[pointcost_df['Player'] == 'Kobe Bryant'].reset_index(drop=True)
bryant_pointcost_df.head()
```

```python
james_pointcost_df = pointcost_df[pointcost_df['Player'] == 'Lebron James'].reset_index(drop=True)
james_pointcost_df.head()
```

```python
jordan_pointcost_df = pointcost_df[pointcost_df['Player'] == 'Michael Jordan'].reset_index(drop=True)
jordan_pointcost_df.head()
```

```python
bryant_points = bryant_pointcost_df['Points'].sum()
bryant_time = bryant_pointcost_df['Minutes Played'].sum()
bryant_salary = bryant_pointcost_df['Salary'].sum()
bryant_adjsalary = bryant_pointcost_df['Adj Salary'].sum()
bryant_career = len(bryant_pointcost_df)

james_points = james_pointcost_df['Points'].sum()
james_time = james_pointcost_df['Minutes Played'].sum()
james_salary = james_pointcost_df['Salary'].sum()
james_adjsalary = james_pointcost_df['Adj Salary'].sum()
james_career = len(james_pointcost_df)

jordan_points = jordan_pointcost_df['Points'].sum()
jordan_time = jordan_pointcost_df['Minutes Played'].sum()
jordan_salary = jordan_pointcost_df['Salary'].sum()
jordan_adjsalary = jordan_pointcost_df['Adj Salary'].sum()
jordan_career = len(jordan_pointcost_df)

print("BRYANT")
print(f"points: {bryant_points}")
print(f"time: {bryant_time}")
print(f"salary: {bryant_salary}")
print(f"salary: {bryant_adjsalary}")
print(f"career length: {bryant_career} seasons")
print("===")
print("JAMES")
print(f"points: {james_points}")
print(f"time: {james_time}")
print(f"salary: {james_salary}")
print(f"salary: {james_adjsalary}")
print(f"career length: {james_career} seasons")
print("===")
print("JORDAN")
print(f"points: {jordan_points}")
print(f"time: {jordan_time}")
print(f"salary: {jordan_salary}")
print(f"salary: {jordan_adjsalary}")
print(f"career length: {james_career} seasons")
```

```python
calc_list = []
def calc(points,time,salary):
    cost_per_point = round((salary/points),2)
    calc_list.append(cost_per_point)
    cost_per_minute = round((salary/time),2)
    calc_list.append(cost_per_minute)
    print(f'cost per point: ${cost_per_point}')
    print(f'cost per minute: ${cost_per_minute}')
    
    
print("CAREER-WIDE AVERAGE COST PER MINUTE")
print("BRYANT")
calc(bryant_points,bryant_time,bryant_salary)
print("===")
print("JAMES")
calc(james_points,james_time,james_salary)
print("===")
print("JORDAN")
calc(jordan_points,jordan_time,jordan_salary)
```

```python
# set width of bars
barWidth = 0.25
 
# set heights of bars
bryant = [calc_list[0], calc_list[1]]
james = [calc_list[2], calc_list[3]]
jordan = [calc_list[4], calc_list[5]]
 
# Set position of bar on X axis
r1 = np.arange(len(bryant))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bryant, color='blue', width=barWidth, edgecolor='white', label='Kobe Bryant')
plt.bar(r2, james, color='orange', width=barWidth, edgecolor='white', label='Lebron James')
plt.bar(r3, jordan, color='purple', width=barWidth, edgecolor='white', label='Michael Jordan')
 
# Add xticks on the middle of the group bars
plt.xlabel('Career-wide Average Cost Per Minute', fontweight='bold')
plt.ylabel('Price in USD', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bryant))], ['Cost Per Point', 'Cost Per Minute Played'])

 
# Create legend & Show graphic
plt.legend()
plt.savefig('images/raw_plot.jpg') 
plt.show()
```
